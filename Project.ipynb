```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histograms for numerical columns
student.hist(figsize=(10, 8), bins=20)
plt.suptitle("Distribution of Numerical Columns")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

bar_width = 0.2
plt.bar(student['GPA'], student['Study_Hours_Per_Day'], color='skyblue', width= bar_width)

# Adding title and labels
plt.title('Student Study Hours per Day by GPA', fontsize=14)
plt.xlabel('GPA', fontsize=12)
plt.ylabel('Study Hours per Day', fontsize=12)

plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Rotate x-axis labels if necessary for better clarity
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
